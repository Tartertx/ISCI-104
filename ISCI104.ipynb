```python
import pandas as pd
import numpy as np

df = pd.read_csv('Isci-recording-1.csv')
df2 = pd.DataFrame()

# df2['Date'] = df['Date']
df2['PM10'] = df['Particulate Matter-PM10 (ug/m&#179;)']
df2['PM2.5'] = df['Particulate Matter-PM2.5 (ug/m&#179;)']
df2['Ozone'] = df['Ozone-O&#x2083; (ppb)']
df2['AQI'] = df['Air Quality Index-AQI ']
df2['TempC'] = df['Internal Temperature-Internal Temperature (&#x2103;)']

for col in df2.columns:
    print(f'The median for {col} is {np.median(df2[col])}')
```
